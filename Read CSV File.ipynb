```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# df = pd.read_csv('100.csv', sep=',', nrows=500)
df = pd.read_csv('100.csv', sep=',', skiprows=0, nrows=500)
datas = np.array(df.values)
times = datas[:, 0]
ml2 = datas[:, 1]
v5 = datas[:, 2]
plt.figure(1, figsize=(16, 3))
plt.plot(times, ml2)
plt.plot(times, v5)
plt.ylabel("mV")
plt.xlabel("Times")
plt.title("ECG Graph")
plt.show()
```
